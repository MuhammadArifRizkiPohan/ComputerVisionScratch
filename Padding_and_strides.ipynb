```python
model = Sequential()

#first layer
#padding (same) = with padding
model.add(Conv2D(32, kernel_size=(3, 3), padding="same", activation="relu", input_shape=(28,28,1)))#input shape only in first layer
#multiple layer
model.add(Conv2D(32, kernel_size=(3, 3), padding="same", activation="relu"))
model.add(Conv2D(28, kernel_size=(3, 3), padding="same", activation="relu"))

#Flatten layer
model.add(Flatten())

#output layer
model.add(Dense(128, activation="relu"))
model.add(Dense(10, activation="softmax"))
```

```python
model.summary()
```

```python
#No padding
model = Sequential()

#first layer
#padding (same) = with padding
model.add(Conv2D(32, kernel_size=(3, 3), padding="valid", activation="relu", input_shape=(28,28,1)))#input shape only in first layer
#multiple layer
model.add(Conv2D(32, kernel_size=(3, 3), padding="valid", activation="relu"))
model.add(Conv2D(28, kernel_size=(3, 3), padding="valid", activation="relu"))

#Flatten layer
model.add(Flatten())

#output layer
model.add(Dense(128, activation="relu"))
model.add(Dense(10, activation="softmax"))
```

```python
model.summary()
```

```python
#No padding wtih stride
model = Sequential()

#first layer
#padding (same) = with padding
model.add(Conv2D(32, kernel_size=(3, 3), padding="valid",strides=(2,2), activation="relu", input_shape=(28,28,1)))#input shape only in first layer
#multiple layer
model.add(Conv2D(32, kernel_size=(3, 3), padding="valid",strides=(2,2), activation="relu"))
model.add(Conv2D(28, kernel_size=(3, 3), padding="valid",strides=(2,2), activation="relu"))

#Flatten layer
model.add(Flatten())

#output layer
model.add(Dense(128, activation="relu"))
model.add(Dense(10, activation="softmax"))
```

## Preprocess `sperm.jpg`

### Subtask:
Load the `sperm.jpg` image, resize it to `(28,28)` pixels, convert it to grayscale, reshape it to match the model's input `(1, 28, 28, 1)`, and normalize pixel values to between 0 and 1.


**Reasoning**:
I will convert the loaded image to grayscale and then resize it to 28x28 pixels as specified in the instructions.



```python
img_gray = img.convert('L')
img_resized = img_gray.resize((28, 28))
print("Grayscale image mode:", img_gray.mode)
print("Resized image size:", img_resized.size)
```

**Reasoning**:
I will convert the resized image to a NumPy array, reshape it, and normalize its pixel values to complete the preprocessing steps.



```python
import numpy as np

img_array = np.array(img_resized)
img_reshaped = img_array.reshape(1, 28, 28, 1)
img_normalized = img_reshaped / 255.0

print("Shape after reshaping:", img_normalized.shape)
print("Pixel value range after normalization: ", img_normalized.min(), "-", img_normalized.max())
```

**Reasoning**:
The preprocessing of the image is complete. Now, I will perform inference using the previously defined CNN model (`model`) on the `img_normalized` data.



```python
predictions = model.predict(img_normalized)
predicted_class = np.argmax(predictions, axis=1)

print("Model predictions:", predictions)
print("Predicted class (index):", predicted_class)
```
